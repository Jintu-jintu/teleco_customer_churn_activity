```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#importing dataset
df=pd.read_excel('D:\ICT kerala\week5\Copy of Telco-Customer-Churn.xlsx')
```

# 1. Compare churn count w.r.to gender

```python
sns.set_style('whitegrid')
sns.countplot(x='Churn',hue='gender',data=df)
```

# 2. Find out how many female senior citizens are there in the dataset

```python
fem_senior=df[(df['gender']=='Female') & (df['SeniorCitizen']==1)]
len(fem_senior)
```

```python
sns.set_style('whitegrid')
sns.catplot(x=df[df['SeniorCitizen']==1]['gender'],kind='count',data=df)
```

# 3. Compare 'tenure' with 'total carges'

```python
# comparing tenure with total charges
sns.relplot(x='tenure',y='TotalCharges',data=df,aspect=2)
```

```python
# As the value of 'tenure' increases the value of 'total charge' also increases
```

# 4. Which contract prefered by senior citizen

```python
sns.set_style('whitegrid')
sns.countplot(x=df[df['SeniorCitizen']==1]['Contract'])
```

```python
# month-to-month is the contract type that is prefered by senior citizens
```

# 5. Findings on payment method

```python
sns.set_style('whitegrid')
sns.catplot(x='PaymentMethod',kind='count',data=df,aspect=2)
```
